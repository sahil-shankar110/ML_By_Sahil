```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import load_diabetes
x,y = load_diabetes(return_X_y=True)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=4)
```

```python
from sklearn.linear_model import SGDRegressor
```

```python
sgd = SGDRegressor(penalty="l2",max_iter=500,learning_rate="constant",eta0=0.1,alpha=0.001)
```

```python
sgd.fit(x_train,y_train)
```

```python
y_pred = sgd.predict(x_test)
```

```python
from sklearn.metrics import r2_score
print(r2_score(y_test,y_pred))
```

```python
print(sgd.coef_)
print(sgd.intercept_)
```

```python
from sklearn.linear_model import Ridge
```

```python
r = Ridge(alpha=0.001,max_iter=500,solver="sparse_cg")
```

```python
r.fit(x_train,y_train)
```

```python
y_pred = r.predict(x_test)
r2_score(y_test,y_pred)
```

```python
print(r.coef_)
print(r.intercept_)
```

```python
class R_GD:
  def __init__ (self,alpha,epochs,l_rate):
    self.alpha = alpha
    self.epochs = epochs
    self.l_rate = l_rate
    self.coef_ = None
    self.intercept_ = None


  def fit(self,x_train,y_train):

    self.coef_ = np.ones(x_train.shape[1])
    self.intercept_ = 0
    w = np.insert(self.coef_,0,self.intercept_)
    x_train = np.insert(x_train,0,1,axis=1)

    for i in range(self.epochs):

      w_der = np.dot(x_train.T,x_train).dot(w) - np.dot(x_train.T,y_train) + (self.alpha * w)
      w = w - (self.l_rate * w_der)

    self.coef_ = w[1:]
    self.intercept_ = w[0]

    print(self.coef_)
    print(self.intercept_)

  def predict(self,x_test):
    return np.dot(x_test,self.coef_) + self.intercept_
```

```python
rgd = R_GD(alpha=0.001,epochs=500,l_rate=0.005)
```

```python
rgd.fit(x_train,y_train)
```

```python
y_pred = rgd.predict(x_test)
r2_score(y_test,y_pred)
```
